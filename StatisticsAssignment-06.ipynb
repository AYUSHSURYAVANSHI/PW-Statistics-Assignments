```python
# #  Q2: Differentiate between one-tailed and two-tailed tests.
# Answer :-
# One-Tailed and Two-Tailed Tests are types of statistical hypothesis tests that differ in their directional focus when making inferences about a population parameter, such as the population mean. Here's how they differ:

# One-Tailed Test:

# Directional Focus: A one-tailed test is used when you have a specific directional hypothesis. In other words, you are interested in determining if a population parameter is greater than or less than a certain value, but not both. It focuses on only one direction of the distribution.
# Critical Region: In a one-tailed test, the critical region (the region of extreme sample values that would lead to the rejection of the null hypothesis) is located in one tail of the distribution. The critical region is set up in the direction specified by the directional hypothesis.
# Hypotheses: The null hypothesis (H0) typically states that there is no significant difference, and the alternative hypothesis (Ha) specifies the direction of the difference.
# Example: If you want to test whether a new drug increases the average reaction time of participants, you would use a one-tailed test with the null hypothesis (H0) that the drug has no effect or decreases reaction time, and the alternative hypothesis (Ha) that the drug increases reaction time.

# Two-Tailed Test:
# Non-Directional Focus: A two-tailed test is used when you have a non-directional hypothesis. It's appropriate when you want to determine if a population parameter is different from a certain value, without specifying whether it's greater or less than that value. It considers both tails of the distribution.
# Critical Region: In a two-tailed test, the critical region is divided into two parts, one in each tail of the distribution, representing extreme values in both directions from the null hypothesis value.
# Hypotheses: The null hypothesis (H0) typically states that there is no significant difference, and the alternative hypothesis (Ha) specifies that the population parameter is different from the hypothesized value, without specifying the direction of the difference.

# Example: If you want to test whether a coin is fair (i.e., has a 50% chance of landing heads), you would use a two-tailed test. The null hypothesis (H0) would be that the coin is fair (p = 0.50), and the alternative hypothesis (Ha) would be that the coin is not fair (p ≠ 0.50).
```

```python
# #  Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for 
# # each type of error.
# Answer :-
# In hypothesis testing, Type 1 and Type 2 errors are two distinct types of mistakes that can occur when making decisions based on statistical tests. These errors are associated with the conclusions drawn from the results of a hypothesis test and can have important implications in various fields, including science, medicine, and quality control.

# Type 1 Error (False Positive):

# Type 1 error, also known as a false positive or alpha error, occurs when you reject a null hypothesis that is actually true. In other words, you conclude that there is a significant effect or difference when there isn't one in reality.
# The probability of making a Type 1 error is denoted by the symbol α (alpha), and it is typically set at a predetermined significance level (e.g., 0.05 or 5%).
# Type 1 errors are more concerning when the consequences of a false positive are severe or costly.
# Example Scenario:
# Imagine a pharmaceutical company conducting clinical trials to test the efficacy of a new drug. The null hypothesis (H0) is that the drug has no effect, while the alternative hypothesis (Ha) is that the drug is effective. If the company incorrectly concludes that the drug is effective (rejecting the null hypothesis) when, in reality, it has no effect, this would be a Type 1 error. Patients may be subjected to a drug with no real benefit and potential side effects.

# Type 2 Error (False Negative):

# Type 2 error, also known as a false negative or beta error, occurs when you fail to reject a null hypothesis that is false. In other words, you conclude that there is no significant effect or difference when one exists.
# The probability of making a Type 2 error is denoted by the symbol β (beta). The complement of β is the statistical power (1 - β), which represents the probability of correctly rejecting a false null hypothesis.
# Type 2 errors are more problematic when failing to detect a real effect has serious consequences, such as in medical diagnostics.
# Example Scenario:
# Suppose a medical test is used to screen for a life-threatening disease. The null hypothesis (H0) in this case is that the individual does not have the disease, and the alternative hypothesis (Ha) is that the individual does have the disease. If the test fails to detect the disease in a patient who actually has it, this would be a Type 2 error. The patient may not receive timely treatment, leading to adverse health outcomes.

# In hypothesis testing, the trade-off between Type 1 and Type 2 errors is managed by selecting an appropriate significance level (α) and designing the study to achieve a desired level of statistical power (1 - β). Researchers must carefully consider the consequences of each type of error and choose the appropriate balance between them based on the specific context and the relative costs of making these errors.
```

```python
# #  Q4:  Explain Bayes's theorem with an example.
# Answer :-
# Bayes' theorem is a fundamental concept in probability theory and statistics that describes how to update the probability of an event based on new evidence or information. It's named after Thomas Bayes, an 18th-century statistician and theologian. Bayes' theorem is particularly useful in scenarios where you want to revise your beliefs about an event in light of new data.

# The theorem can be expressed as follows:
# P(A∣B)=P(B∣A)⋅P(A)/p(B)
  
#   Where:
# P(A∣B) is the posterior probability of event A given that event B has occurred.

# P(B∣A) is the likelihood of event B occurring given that event A has occurred.

# P(A) is the prior probability of event A, which is your initial belief in the probability of A.

# P(B) is the probability of event B, which acts as a normalization constant.

# To illustrate Bayes' theorem with an example, let's consider a medical diagnosis scenario:

# Suppose you want to determine the probability that a patient has a particular disease, let's call it Disease X. You have some initial information about the patient's risk factors and some test results. Here's how Bayes' theorem can be applied:

# P(A): Prior probability - This is your initial belief about the probability of the patient having Disease X based on their risk factors. Let's say you estimate it to be 5%, so P(A)=0.05.
# P(B∣A): Likelihood - This is the probability of observing the test results you have if the patient indeed has Disease X. Let's say the test is 95% accurate, so P(B∣A)=0.95.

# P(B∣¬A): The likelihood of observing the same test results if the patient does not have Disease X. In this case, it might be 5% false-positive rate, so P(B∣¬A)=0.05.

# P(B): The total probability of getting the observed test results, which can be calculated using the law of total probability:

# P(B)=P(B∣A)⋅P(A)+P(B∣¬A)⋅P(¬A)

# Let's assume that 
# P(¬A) is 1 - P(A)=0.95.

# Now, using Bayes' theorem, you can calculate the posterior probability 

# P(A∣B), which is the updated probability of the patient having Disease X given the test results.

# P(A∣B)= P(B∣A)⋅P(A)/P(B)
#     =0.95*0.05/0.95*0.05+0.05*0.95
#     =0.0475/0.0475+0.0475=0.5

# So, after considering the test results, the probability that the patient has Disease X is 50%. This is an example of how Bayes' theorem can be used to update your beliefs with new evidence or information, making it a valuable tool in fields such as medicine, machine learning, and more.
```

```python
# # Q6. Use Bayes' Theorem to calculate the probability of an event occurring given prior knowledge of the 
# # event's probability and new evidence. Provide a sample problem and solution.
# Answer :-
# Monty Hall Problem:
# You are a contestant on a game show. The host, Monty Hall, presents you with three doors. Behind one of the doors is a car, and behind the other two doors are goats. You choose one of the doors. Monty, who knows what is behind each door, then opens one of the other two doors to reveal a goat. You are given the opportunity to switch your choice to the remaining unopened door. Should you switch, or stick with your initial choice, to maximize your chances of winning the car?

# Let's use Bayes' theorem to calculate the probability of winning the car if you switch your choice.

# Solution:
# To apply Bayes' theorem, let's define the following events:

# A: You initially choose the door with the car behind it.

# B: Monty opens one of the other two doors, revealing a goat.
# We want to find P(A∣B), which is the probability that you initially chose the car (event A) given that Monty revealed a goat (event B).

# First, let's establish some prior probabilities:

# P(A): The probability of initially choosing the car is 1/3 because there are three doors, and you choose one randomly.

# P(B∣A): The probability that Monty reveals a goat if you initially chose the car is 1 because he has no choice but to reveal a goat behind one of the other two doors.

# P(B∣¬A): The probability that Monty reveals a goat if you initially chose a goat is 1 because Monty again has no choice but to reveal the other goat behind the remaining door.
# Now, we can use Bayes' theorem to calculate P(A∣B):

# P(A∣B)= P(B∣A)⋅P(A)/P(B)

# To calculate P(B), we'll use the law of total probability:

# P(B)=P(B∣A)⋅P(A)+P(B∣¬A)⋅P(¬A)

# P(¬A): The probability of initially choosing a goat is 2/3 because there are two goats and three doors.
# Now, we can plug these values into the formula:

# P(A∣B)= 1*1/3/1*1/3 + 1*2/3 = 1/3/1/3+2/3 = 1/3
# So, the probability of winning the car if you switch your choice (given that Monty revealed a goat) is 1/3, while the probability of winning if you stick with your initial choice is 2/3. This result might be counterintuitive, but it demonstrates that you should indeed switch doors to maximize your chances of winning the car in the Monty Hall problem.
```
